```python
# 도수
# 히스토그램

df_tips = sns.load_dataset('tips')

f = plt.figure()

ax = f.add_subplot(1,1,1)
ax.hist(df_tips['smoker'],color = 'salmon')
ax.set_xlabel('Smoker')
ax.set_ylabel('Tip')
ax.set_title('Smoker and tip')
```

```python
# 산점도

df_tips = sns.load_dataset('tips')

f = plt.figure()
ax = f.add_subplot(1,1,1)

ax.scatter(df_tips['total_bill'], 
           df_tips['tip'],
          color = 'salmon',
)

ax.set_xlabel('Total_bill')
ax.set_ylabel('Tip')
ax.set_title('Total bill and tip')
```

```python
df_tips = sns.load_dataset('tips')

f = plt.figure()

ax = f.add_subplot(1,1,1)

ax.boxplot([
    df_tips[df_tips ['sex'] == 'Female']['tip'],
    df_tips[df_tips['sex'] == 'Male']['tip']
], labels = ['Female', 'Male'],
   
)


ax.set_xlabel('Sex')
ax.set_ylabel('Tip')
ax.set_title('Boxplot of Tips by Sex')
```

```python
df_tips = sns.load_dataset('tips')

f, ax = plt.subplots()
ax = sns.distplot(df_tips['total_bill'], rug = True, 
                  color = 'salmon')

ax.set_title('Total bill histogram density and rug plot')
ax.set_ylabel('Tip')
ax.set_xlabel('Total_bill')
```

```python
df_tips = sns.load_dataset('tips')

f,ax = plt.subplots()
ax = sns.kdeplot(data = df_tips['total_bill'], data2 = df_tips['tip'], 
                 shade = True, color = 'black')
ax.set_title('Kernel density plot of total bill and tip')
ax.set_xlabel('Total_bil')
ax.set_ylabel('Tip')
```
