```python
# ebola_long.info()

# ebola_long.variable.str  # 데이터 프레임이 갖고 있는 컬럼 중 문자열 타입이 오브젝트인 것에는 .str 속성이 있음.
# 찍어보니 타입은 stringmethod임
# 문자열 처리 메소드가 잔뜩 들어있는 ..
# dir(ebola_long.variable.str)
ebola_long.variable.str.split('_')
```

```python
variable_split = ebola_long.variable.str.split('_')  # ebola 데이터가 갖고있는 variable 이라는 것에서 구분자. _로 구분하라. 
print(variable_split[:5]) # 이 시리즈의 데이터 타입은 리스트임 ! 
# 아랫줄 코드에선 여기 데이터를 끄집어 낸다는 소리 ! 시리즈도 str 속성이 있기 때문에 이거 끄집어내는 get사용함.
```

```python
status_values = variable_split.str.get(0) # case 또는 death라 하는 단어 들어감.
country_values = variable_split.str.get(1) # 두번째 요소를 꺼냈으니, 국가명이 들어감.
# 이거 두개 타입 확인하면 이것들도 시리즈로 나옴.
# get() 안에 숫자는 인덱스 번호만 늘려가면 됨 !

print(status_values[:5])
```

```python
print(country_values)
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
print(ebola_long.head())
```

# 기상 데이터의 여러 열을 하나로 정리하기

```python
weather = pd.read_csv('../doit_pandas/data/weather.csv') 
weather.info()

print(weather.iloc[:5, :11]) # 전체 중 5개, 전체 컬럼 중 11개만 보기
# element는 범주형, category일 가능성이 큼.
# id, year, month, element 는 연속형 변수 아님.
```

```python
# melt를 이용해서 길게 늘어져 있는 데이터를 보기 쉽게 변환
weather_melt = pd.melt(
    weather, 
    id_vars=['id', 'year', 'month', 'element'], 
    var_name='day', 
    value_name='temp'
) 

print(weather_melt.head())
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'], # 여기까지 고정
    columns='element', # 피벗할 열 이름 (element에 있는 값으로 새롭게 열이 생김.) (유형별로 쪼개짐.)
    values='temp' # 새롭게 생긴 열에 들어갈 데이터를 공급. 
)

# print(weather_tidy)
# weather_tidy
weather_tidy.reset_index() # 이게 보기 편함.

# weather_tidy.index # 피벗하게 되면 고정시키로 한 컬럼들이 인덱스로 바뀜 ! .. ( mx, 2010, 1, d30 ) 이게 하나의 인덱스
# weather_tidy.columns # tmax, tmin 두개만 실제 데이터프레임의 컬럼이 됨 
```

# 빌보드 차트의 중복 데이터 처리하기

```python
billboard = pd.read_csv('../doit_pandas/data/billboard.csv')
billboard.info()

billboard_long = pd.melt(
    billboard, 
    id_vars=['year', 'artist', 'track', 'time', 'date.entered'], 
    var_name='week',  # 반복적으로 이어지는 week를 회전. (새로운 열의 이름 지정)
    value_name='rating'
)

print(billboard_long.shape) # 차원정보는 당연히 행이 늘어남. (늘어져있던 컬럼을 행을 길게 바꿔서)

print(billboard_long.head())
```

```python
print(billboard_long[billboard_long.track == 'Loser'].head())
# 원래 이거 열 뽑아낼 때 쓰는건데 지금 행 뽑아내는데 사용.
# 데이터 프레임에 [] 열면 컬럼 뽑아내는건데, 이건 불린조건 사용해서 행 뽑아냄.
# loc나 iloc 쓰지 않고 이거 써서 뽑아내는 경우가 많음 !! 인덱스 잘 사용하지 않음..! 
# SQL의 WHERE 처럼 ! 
```

```python
billboard_songs = billboard_long[
    ['year', 'artist', 'track', 'time']
] 

print(billboard_songs.shape)
print(billboard_songs.head())
```

```python
billboard_songs = billboard_songs.drop_duplicates()  # 중복된 행을 삭제하라

print(billboard_songs.shape)
```

```python
# billboard_songs['id'] = range(len(billboard_songs)) 
billboard_songs['id'] = range(billboard_songs.shape[0]) # shape 속성의 첫번째 요소를 사용 !! 
# 위에 shape 튜플의 첫번째 요소와 len 과 같음.
# 0부터 317까지 숫자를 지정하라.
# 현재 중복된 행 삭제한거니까 인조키 생성하는 것과 같음 ! 인덱스의 값과 동일 ! 
print(billboard_songs.head(n=10))
```

```python
# DataFrame.merge() method -> Equi Join
billboard_ratings = billboard_long.merge( 
    billboard_songs, on=['year', 'artist', 'track', 'time']
) # 두 개의 데이터프레임을 join.. 
  # melt 한 long 데이터와 중복을 제거한 songs 데이터 프레임을 왜 merge 했는가 ??
  # 
print(billboard_ratings.shape)
```

```python
print(billboard_ratings.head())
# 중복된 데이터 없이 나왔으니 natural 조인과 같음.
print('-'*80)
print(billboard_long.head()) # 위아래 데이터 비교해보면 id컬럼만 빼면 비슷.. 
# 그럼 왜 merge 한건지 ??

print(billboard_ratings.shape)
```

# 뉴욕 택시 데이터 준비

```python
import os 
import urllib.request 

# 네트워크 상태에 따라 5 ~ 15분이 소요됩니다.
# f = open~~
# data = f.read()
# f.close() # 이렇게 하던것과 같음 !!

with open('../doit_pandas/data/raw_data_urls.txt', 'r') as data_urls:
#     for lineno, line in enumerate(data_urls): # enumerate 씌우면 한개 데이터만 가져오는것 뿐만 아니라 번호도
#                                                                     # 튜플형태로 가져올 수 있음.
#         print(lineno, line)
#         pass
    for line, url in enumerate(data_urls): # url 주소를 하나씩 읽어다가~
        if line == 5:
            break 
        fn = url.split('/')[-1].strip() # url에서 받아온 걸 / 로 쪼갬..
        fp = os.path.join('', '../doit_pandas/data', fn)
        print(url)
        print(fp)
        urllib.request.urlretrieve(url, fp) # 브라우저에서 요청할 url, 받아온 데이터를 파일로 쌓도록 파일에 대한 정보
                                            # fp를 넣으면 받아온 데이터를 쌓을 수 있음.
```

```python
## 여기는 위에 단락 이해하기 위해 예제 실행!!! 모두 주석처리할거임!

import os 
import urllib.request 

# 네트워크 상태에 따라 5 ~ 15분이 소요됩니다.
# f = open~~
# data = f.read()
# f.close() # 이렇게 하던것과 같음 !!

with open('../doit_pandas/data/raw_data_urls.txt', 'r') as f:
    for index, line in enumerate(f):
        print('- index: {}, line: {}'.format(index, line)) # 지정된 파일에서 하나씩 읽어서 행번호와 줄 저장하고
        pass
        
        if index ==2:
            break
            pass
        
        filename = line.split('/')[-1].strip() # 우리가 읽은 line을 쪼개서 담을것임.
        print('\t+ filename :', filename) # split 메소드는 구분자로 쪼갠 뒤 담긴 결과를 리스트로 반환. 
        # 맨 마지막 토큰은 csv 파일명으로 나왔음 ! 
        # 파일명만 얻기 위해서 -1 인덱스를 사용 !! 리스트 중 마지막 요소 파일명 얻음.
        # 그 후 진행한 strip은 완전한 파일 이름만 나오도록 ㅂ공백 없애줌
        
        import os
        filepath = os.path.join('', '../doit_pandas/data', filename) # 어디다 저장할지
        # 파일 위치 이름 저장 !! 
        print('\t+ filepath :', filepath)
        
        urllib.request.urlretrieve(line, filepath) # 웹브라우저가 받아온 문서를 우리가 저장한 경로에 저장.
        print('- {} get completed.'.format(filename)) # 매번 하나의 파일을 받아올때마다 완료 했다고 문자 표시
        
        pass
        
```
