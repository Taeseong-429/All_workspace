```python
# 위 그림을 오른쪽으로 쓰러뜨린다고 생각해보면, Female이 지불한 가장 작은 팁과 가장 큰 팁은 왼쪽 끝, 오른쪽 끝에 그려짐. (Bar형태로 !) 
# 박스 위 아래 선을 수염이라고 부름.. 위쪽에 있는 수염을 윗수염, 아랫수염 ? 이라고 함.
# 박스에서 맨 아랫부분을 전체 값의 1분위수 라고 함.
# 박스 안의 직선은 중위수 라고 부름. 전체 4분위 중 50% 이기 때문에 제 2분위수.
# 박스의 맨 윗부분을 전체 데이터의 3분위수라고 함. 약 4달라에 해당하는 값이 전체 데이터의 3분위 수
# 25%, 50%, 75%까지의 데이터가 박스 안에 있다는 얘기니까, 대부분의 데이터가 여기에 다 속한다는 소리.
# 1분위수와 3분위수 사이의 범위를 IQR 이라고함. Interquartile Range
# 이상치는 미니멈, 맥시멈 바깥에서 나옴. 대부분 맥시멈 바깥에서 나옴.
```

```python
# 위 그래프에선 남자가 팁을 좀 더 지불함. 
# IQR의 범위가 여자보다 남자가 약간 길게 되어있음. 이 범위대에 지불하는 수는 남자가 약간 더 많다는 뜻.
# 분포의 정점.. 중위수 선이 남자가 좀 더 높기 때문에 약간 더 높은 금액의 팁을 지불한다.
# max 위쪽에 있는 이상치는 남자가 더 많음. 이상치 중의 극단치 .. 가 많음. 이걸 제거하고 평균을 구해야함. 결측치로 만들기 ! 
```

# 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else :
        return 1
# 매개변수의 이름인 성별로 설정하고, 성별이 들어오면 값을 되돌려줌.
# 성별은 범주형 변수임. 남자이면 1, 여자면 0으로 바꿔줌. 사용법은 아래에 나옴
```

```python
tips = sns.load_dataset('tips')
print(tips.head())

print('-'*80)

tips['sex_color'] = 1 # 새로운 열을 추가하고 값을 1로 ! 
print(tips.head())

# apply() metohd : 인자로 전달된 함수를 Broadcasting. 하는것임.
# apply 왼쪽에 있는 시리즈의 값들이 apply 에 들어가 매개변수로 작용한다는 의미
# 여러개의 데이터를 함수로 적용시키는.. 그런 방법인듯 !! 작동 순서는 왼쪽의 값들이 ㅇ오른쪽 함수의 매개변수로 들어가서 작동
tips['sex_color'] = tips['sex'].apply( recode_sex )  
# 이렇게 함수 적용시키면 성별데이터의 값이 함수에 매개변수로 들어가서 값을 다시 반환해줌 ! 

tips.head()
```

```python
f = plt.figure()

ax = f.add_subplot(1, 1, 1)

ax.scatter (
    x = tips['total_bill'], # x 축에는 이거
    y = tips['tip'], # y 축에는 이거.. 이렇게 입력. 
    s = tips['size'] * 10, # size # 사이즈를 조절하는 인자. 
    c = tips['sex_color'], # color : 데이터포인트의 색깔까지 구분하자 ! apply 메소드로 만든 새로운 변수를.(0과 1로 구분된거) 
    alpha = 1.0            # alpha - transparency  투명도 : 알파값이 낮으면 낮을수록 투명해짐. # 이거 RGBA 관련 용어 정의는 노션에 메모.
) 

ax.set_title('Total Bill vs Tip Color')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

dir(ax) # 좌표계 ax 개체가 주어졌을 때 내가 쓸 수 있는 메소드의 종류를 확인 가능
# ax
```

```python
ax = plt.subplots() 

ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)  # 회귀선을 뺴고 그릴 수 있음.
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

```python
jg = sns.jointplot( x = 'total_bill', y = 'tip', data = tips) 
# 타입이 그리드로 끝나는 객체가 리턴됨. grid(격자). 최소 한개 이상의 그래프를 그리겠다. 
# 이것도 명시적으로 figure 객체를 튜플로 제공할 것 같음.. 타입 말하는건가

jg.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')

jg.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03 )
# 전혀 다른 plot 두개를 결합시켜서 표시
# 위에는 히스토그램, 아래는 산점도 두개를 결합시켜서 보여줌. 
# 총 지불액의 히스토그램은 위에, tip의 히스토그램은 오른쪽에 ? (변수 한개이기 때문에)
# 히스토그램은 빈도수를 보여주기 때문에 사실 따로 그려주는게 좋음.(산점도와)
```

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") 
# 산점도를 나타내는 데이터 포인트가 육각형으로 바뀜. hexbin
# 점의 kind 라는 의미 ! 포인트를 점이 아니라 육각형으로 보여준다는 뜻. 육각형 하나하나가 밀도까지 보여줌.. 
# 겹치는 부분은 색깔로 드러나기 때문에 밀도로 표현 가능해짐 ! 
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
iris = sns.load_dataset('iris')

g = sns.PairGrid(iris, hue = "species") # 격자 구조.. # 범례를 어디에 지정할지 ! # hue 는 색상을 씌우는 것.
# 여기서 hue 를 sepcies 종류에 입히겠다 라는 의미 !
# 품종별로 색을 입혀서 산점도를 그림 ! 

g = g.map_diag(plt.hist) # 대각선 - diagonal : 여기 대각선에 그릴 그림을 hist로 한다는 얘기.
g = g.map_offdiag(plt.scatter)

g = g.add_legend() # legend : 그래프 상의 범례를 만드는 것.
```

```python
iris = sns.load_dataset('iris')

g = sns.PairGrid(iris, hue = "species") # 격자 구조.. # 범례를 어디에 지정할지 ! # hue 는 색상을 씌우는 것.
# 여기서 hue 를 sepcies 종류에 입히겠다 라는 의미 !
# 품종별로 색을 입혀서 산점도를 그림 ! 

g = g.map_diag(plt.hist, histtype="step", linewidth = 3) # hist 타입을 step로..
g = g.map_offdiag(plt.scatter)

g = g.add_legend() # legend : 그래프 상의 범례를 만드는 것.
```

```python
iris = sns.load_dataset('iris')
iris.head()

g = sns.PairGrid(iris)

g = g.map_diag(plt.hist, edgecolor = 'w') # w는 white의 준말. (edgecolor : 가장자리의 색을 white로 준다)
# 겹치는걸 명확하게 확인하기 좋을 때 사용하나 ?
# 비슷한 양을 갖고 있는 데이터는 동일한 위치라 구분하기 힘듬. 구분하기 쉽게 할 때 사용 !! 
g = g.map_offdiag(plt.scatter, edgecolor='w', s = 40)
```
