```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../doit_pandas/data/gapminder.tsv', sep = '\t')
df
```

## 문제 1

```python
df_afg = df[ df['country'] == 'Afghanistan' ]
print(df_afg)
df_afg2 = df_afg[['country', 'year', 'pop']]
print(df_afg2)
df_afg2.plot(x = 'year', y ='pop')
plt.title('Evolution of population in Afghanistan')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
```

```python
1. 시간이 지날수록 인구가 증가
2. 1977년 ~ 1982년 사이에 어떠한 사건으로 인해 인구가 감소. special event 발생한 것으로 추정 ex) 전쟁, 이민, 전염병 등등..
3. 1990년대 이후로 인구 증가 기울기가 상대적으로 증가(feat 베이비붐)
4. 1970년 후반부터 1980년 후반까지 전쟁이 지속됨
```

```python
df_afg = df[ df['country'] == 'Afghanistan' ]
print(df_afg)
df_afg2 = df_afg[['country', 'year', 'gdpPercap']]
print(df_afg2)
df_afg2.plot(x = 'year', y ='gdpPercap')
plt.title('Evolution of gdpPercap in Afghanistan')
plt.xlabel('Year')
plt.ylabel('GdpPercap')
plt.grid(True)
```

## 문제 4

```python
df.head(20)
```

```python
df_pivot = df.pivot_table(values='lifeExp',columns='continent' ,index='year')
print(df_pivot)
df_pivot.plot()
plt.title('Evolution population in Afghanistan')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Evolution of lifeExp by continent')
plt.xlabel('Year')
plt.ylabel('LifeExp')
plt.grid(True)
plt.legend(bbox_to_anchor=(0.8, 0., 0.5, 0.7))
plt.axis([1950, 2010, 30, 90])
```

```python
1. 시간이 지날수록 모든 대륙의 평균 수명이 지속적으로 상승
2. 대륙별 평균 수명의 차이를 육안으로 확인 가능
3. 아시아의 평균기대수명 증가폭이 타 대륙보다 높은 것을 확인 가능
4. 아프리카의 평균 기대수명 증가폭이 낮으며, 1990년 ~ 2000년 사이에 증가폭이 가장 낮은 것을 확인가능
5. 오세아니아대륙은 2개의 나라밖에 집계가 되지 않았음 
```
