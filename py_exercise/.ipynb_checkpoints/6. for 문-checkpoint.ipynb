```python
marks = [90, 25, 67, 45, 80]
number = 0

for i in marks :
    number += 1
    if i > 60 :
        print("{}번 학생 : 합격".format(number))
        pass
    else :
        print("{}번 학생 : 불합격".format(number))
        pass
    pass
```

```python
mark = {
        1:90, 
        2:25, 
        3:67, 
        4:45, 
        5:80
}

for i in mark :
    if mark[i] > 60:
        print('{}번 학생 합격'.format(i))
        pass
    else :
        print('{}번 학생 불합격'.format(i))
    pass
```

```python
marks = {
    1 : 90,
    2 : 25,
    3 : 67,
    4 : 45, 
    5 : 80
}

for i in marks :
    if marks[i] > 60 :
        print('{}번 학생 합격입니다.'.format(i))
        pass
    else :
        print('{}번 학생 불합격.'.format(i))
        pass
    pass
```

```python
marks = [90, 25, 67, 45, 80]
number = 0

for i in marks :
    number += 1
    if i > 60 :
        print('{}번 학생 합격입니다.'.format(number))
        pass
    else :
        continue
        pass
    pass
```

```python
marks = [90, 25, 67, 45, 80]
number = 0

for i in marks :
    number += 1
    if i < 60 :
        continue
        pass
    print(f'{number}번 학생 축하합니다. 합격입니다.')
    pass
```

```python
marks = [90, 25, 67, 45, 80]
number = 0

for i in range(len(marks)) :
    number += 1
    if marks[i] < 60 :
        continue
        pass
    print(f'{number}번 학생 합격입니다.')
    pass
```
